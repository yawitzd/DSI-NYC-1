```python
# reserve the first cell, to import all dependencies (python classes) that you will need

import statsmodels.api as sm
import pandas as pd
import pylab as pl
import numpy as np
```

```python
def read_file(path): # path = dynamic input parameter
    return pd.read_csv(path, header = None)
```

```python
# use function to read in cancer Dataset. 
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
df = read_file(path)
```

```python
# find data dictionary of df
df.dtypes
```

```python
df.head(10)
# You want to predict M which is at index 1
```

```python
data_df   = df[list(df.columns[2:12])]
target_df = df[1]
```

```python
data_df
```

```python
target_df
```

```python
#Convert DataFrame to Numpy Array
x = data_df.as_matrix(columns=None)
y = target_df.as_matrix(columns=None)
```

```python
### specify the model
### we are doing a MNLogity aka multinomial logit model because of the M(malignant) and B(benign) in the first column
mdl = sm.MNLogit(y, x)
mdl_fit = mdl.fit()
```

```python
### print model summary ###
mdl_fit.summary()
```

```python
target_df2 = target_df.map([['M']:1],[['B']:0])
```

```python
### aic and bic ###
print(mdl_fit.aic)
print(mdl_fit.bic)
```

```python
### marginal effects ###
mdl_margeff = mdl_fit.get_margeff()
print( mdl_margeff.summary() )
```
